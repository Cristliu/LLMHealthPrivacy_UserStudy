## Statistical Analysis---RQ2---Expectation

### Sign test_central tendency

```python
CI_data = copy.deepcopy(Analysis_df)

new_group_columns = ["the user's basic information", "the user's lifestyle", "the user's mental health information", "the user's medical information"]

score_columns = [
    'if the user has given consent', 
    'if the user is notified before collection',
    'if information is kept confidential and not stored', 'if information will be stored',
    'if information is used for health monitoring',
    'if information is shared with others',
    'if it complies with information protection regulations',
    'if information is provided for academic research',
    'if information is used in public health surveys',
    'if used for data analysis and service improvement'
]

nullprinc_map = {
    "the user's family members": "the user's family members_null",
    'healthcare providers': 'healthcare providers_null',
    'government agencies': 'government agencies_null',
    'LLM service providers': 'LLM service providers_null'
}

results = []


for group in new_group_columns:
    # Sets the mask for selecting the row with the current group column value of 1.
    mask = CI_data[group].astype(int) == 1
    
    for original_group, null_column in nullprinc_map.items():
        # Find rows where original_group is also 1
        combined_mask = mask & (CI_data[original_group].astype(int) == 1)
        selected_columns = score_columns + [null_column]
        group_scores = CI_data.loc[combined_mask, selected_columns].astype(float)
        
        # Add a new null column and set it to the average score of the corresponding null columns
        group_scores['null'] = group_scores[null_column]
        group_scores = group_scores.drop(null_column, axis=1)
        
        # Move the null column to the position of the first column
        columns_order = ['null'] + score_columns
        group_scores = group_scores[columns_order]
        
        group_scores = group_scores.dropna(axis=1, how='all')
        group_scores = group_scores.reset_index(drop=True)
        
        # Perform a two-sided sign test for each column
        for col in group_scores.columns:
            # Counts greater than 0, less than 0, and equal to 0
            positives = (group_scores[col] > 0).sum()
            negatives = (group_scores[col] < 0).sum()
            zeros = (group_scores[col] == 0).sum()
            total_non_zeros = positives + negatives  # Ignore values equal to 0 for testing
            
            # Two-sided sign test, assuming that the median is equal to 0
            if total_non_zeros > 0:
                p_value = binomtest(positives, n=total_non_zeros, p=0.5, alternative='two-sided').pvalue
            else:
                p_value = 1  # If there is no valid sample, the p-value is 1

            bonferroni_alpha = 0.05 / 144
            is_significant = p_value < bonferroni_alpha
            
            # Determining concentration trends
            if is_significant:
                trend = "greater" if positives > negatives else "lesser"
            else:
                trend = "equal"

            results.append({
                'data': group,
                'recipient': original_group,
                'principle': col,
                'Positives': positives,
                'Negatives': negatives,
                'Zeros': zeros,
                'P-Value': p_value,
                'Significant': is_significant,
                'Trend': trend  
            })

results_df = pd.DataFrame(results)
results_df
```

### p-value calculation_1===null VS non-null i.e. comparing p-values of different information flows

```python
CI_data = copy.deepcopy(Analysis_df)


# Define grouped columns and corresponding null columns
new_group_columns = ["the user's basic information","the user's lifestyle","the user's mental health information", "the user's medical information"]

score_columns = [
    'if the user has given consent', 
    'if the user is notified before collection',
    'if information is kept confidential and not stored', 'if information will be stored',
    'if information is used for health monitoring',
    'if information is shared with others',
    'if it complies with information protection regulations',
    'if information is provided for academic research',
    'if information is used in public health surveys',
    'if used for data analysis and service improvement'
]

nullprinc_map = {
    "the user's family members": "the user's family members_null",
    'healthcare providers': 'healthcare providers_null',
    'government agencies': 'government agencies_null',
    'LLM service providers': 'LLM service providers_null'
}


# Create an empty DataFrame to store the scores for each grouping
grouped_scores = pd.DataFrame()
results = []


for group in new_group_columns:
    # Sets the mask for selecting the row with the current group column value of 1.
    mask = CI_data[group].astype(int) == 1
    
    for original_group, null_column in nullprinc_map.items():
        # Find rows where original_group is also 1
        combined_mask = mask & (CI_data[original_group].astype(int) == 1)
        selected_columns =  score_columns + [null_column]
        group_scores = CI_data.loc[combined_mask, selected_columns].astype(float)
        
        # Add a new null column and set it to the average score of the corresponding null columns
        group_scores['null'] = group_scores[null_column]
        group_scores = group_scores.drop(null_column,axis=1)
        
        # Move the null column to the position of the first column
        columns_order = ['null'] + score_columns
        group_scores = group_scores[columns_order]
        
        group_scores = group_scores.dropna(axis=1, how='all')
        group_scores = group_scores.reset_index(drop=True)
        
        # Get the first column as the baseline
        baseline = group_scores.iloc[:, 0]        

        # Compare the baseline column to every other column
        for col in group_scores.columns[1:]:
            # Wilcoxon
            stat, p_value = stats.wilcoxon(baseline, group_scores[col], mode='approx')
            # Bonferroni
            bonferroni_alpha = 0.05 / 128
            is_significant = p_value < bonferroni_alpha         

            results.append({
                'data':group,
                'recipient':original_group,
                'Comparison': col,
                'Statistic': stat,
                'P-Value': p_value,
                'Significant': is_significant,
            })
            
results_df = pd.DataFrame(results)
results_df
```

```python
group_columns = ["the user's family members", 'healthcare providers', "government agencies", 'LLM service providers']

grouped_scores = pd.DataFrame()

for group in group_columns:

    mask = HotMap_data[group].astype(int) == 1
    selected_columns = score_columns + [nullprinc_map[group]]
    group_data = HotMap_data.loc[mask, selected_columns].astype(float)
    
    # Count the number of positive and negative values in each score column (including the null column)
    positive_counts = (group_data[selected_columns] > 0).sum()
    negative_counts = (group_data[selected_columns] < 0).sum()
    zero_counts = (group_data[selected_columns] == 0).sum()
    
    positive_ratios = positive_counts / (positive_counts + negative_counts+zero_counts)
    negative_ratios = negative_counts / (positive_counts + negative_counts+zero_counts)
    zero_ratios = zero_counts / (positive_counts + negative_counts+zero_counts)
    
    # Format for splicing number of positive/negative values, positive ratio/negative ratio
    formatted_scores = positive_counts.astype(str) + '/' + negative_counts.astype(str) + '/' + zero_counts.astype(str) + ', ' + \
                       positive_ratios.round(4).fillna(0).astype(str) + '/' + negative_ratios.round(4).fillna(0).astype(str) + '/' + zero_ratios.round(4).fillna(0).astype(str)
    
    
    formatted_scores['null'] = formatted_scores[nullprinc_map[group]]
    formatted_scores = formatted_scores.drop(nullprinc_map[group])
    
    columns_order = ['null'] + score_columns
    formatted_scores = formatted_scores[columns_order]
    
    grouped_scores[group] = formatted_scores


grouped_scores = grouped_scores.T
grouped_scores
```

```python
results = []

for row in grouped_scores.itertuples(index=True):
    total_positive = 0
    total_negative = 0
    total_zero = 0
    
    for cell in row[1:]:
        # Segmentation of positive/negative/zero
        pos_neg_zero = cell.split(",")[0]
        pos, neg, zero = map(int, pos_neg_zero.split("/"))
        
        total_positive += pos
        total_negative += neg
        total_zero += zero

    total = total_positive + total_negative + total_zero
    positive_ratio = total_positive / total if total != 0 else 0
    negative_ratio = total_negative / total if total != 0 else 0
    zero_ratio = total_zero / total if total != 0 else 0
    
    results.append([total_positive, total_negative, total_zero, 
                    round(positive_ratio, 4), round(negative_ratio, 4), round(zero_ratio, 4)])

summary_df = pd.DataFrame(results, 
                          columns=['Total Positive', 'Total Negative', 'Total Zero', 
                                   'Positive Ratio', 'Negative Ratio', 'Zero Ratio'], 
                          index=group_columns)

summary_df
```
